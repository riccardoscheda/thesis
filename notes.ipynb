```python
import networkx as nx
import numpy as np
import pylab as plt
```

```python
a = np.random.randint(0,2,size=(10,10))
g=nx.from_numpy_matrix(a)
pos = nx.layout.spring_layout(g)
```

```python
nx.draw_networkx(g,pos = pos, with_labels = True)
```

```python
b = np.random.randint(0,2,size=(10,10))
h=nx.from_numpy_matrix(b)

gen = ((i, i+10) for i in range(10))
mapping = dict(gen)
h = nx.relabel_nodes(h, mapping)
hpos = nx.layout.spring_layout(h)
for el in hpos:
    hpos[el][0] = hpos[el][0] + 5
    
nx.draw_networkx(h,pos = hpos, with_labels = True)
nx.draw_networkx(g,pos = pos, with_labels = True)
```

```python
tot = nx.union(g,h)
nx.draw_networkx(tot, with_labels=True)
```

```python
m = np.array(nx.to_numpy_matrix(tot))

plt.imshow(m)
```

```python
m[4][15] = 1
m[18][7] = 1
plt.imshow(m)
```

```python
new = nx.from_numpy_matrix(m)
```

```python
nx.draw_networkx(new, with_labels=True)
```

```python
n = 10
m = 5
a = np.random.randint(0,2,size = (n,n))
b = np.random.randint(0,2,size = (m,m))
plt.imshow(a)
```

```python
tot = np.block([[a,               np.zeros((n, m))],
    [np.zeros((m, n)), b               ]])
plt.imshow(tot)
```

```python
t = nx.from_numpy_matrix(tot)
pos = nx.layout.spring_layout(t)
nx.draw(t,pos = pos)
nx.draw_networkx_edges(t,pos = pos,width = 2,edgelist=[(4,12)],edge_color = "red")
```
